```python
## 最大値を求めるプログラム
def findMax(A, l, r):
    m = int((l + r) / 2) # divide
    if l == r - 1: # 要素数が1個しか無い場合
        return A[l]
    else:
        u = findMax(A, l, m) # 前半部分のSolve
        v = findMax(A, m, r) # 後半部分のSolve
        x = max(u, v) # Conquer
    return x
```

```python
A = [1, 4, 6, 2, 5, 8, 4, 5, 6]
findMax(A, 2, 4)
```

```python
def solve(i, m):
    if m == 0:
        return True
    if i >= n:
        return False
    res = solve(i + 1, m) or solve(i + 1, m - A[i])
    return res
    
```

```python
P = [[0,0], [100,0]]
n = 6
for i in range(n):
    #print(i, P)
    P = sub3(P)
```
